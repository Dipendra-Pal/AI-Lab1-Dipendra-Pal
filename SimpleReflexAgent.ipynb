```python
import random

class VacuumReflexAgent:
    def __init__(self, size=(5, 5)):
        self.size = size
        self.grid = [[random.choice(['C', 'D']) for _ in range(size[1])] for _ in range(size[0])]
        self.pos = [0, 0]
        self.cleaned = 0

    def is_room_clean(self):
        """Check if all cells are clean."""
        return all('D' not in row for row in self.grid)

    def act(self):
        """Perform action based on current cell state."""
        x, y = self.pos
        if self.grid[x][y] == 'D':
            self.grid[x][y] = 'C'
            self.cleaned += 1
            return f"Cleaned ({x},{y})"

        moves = [(x + dx, y + dy) for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]
                 if 0 <= x + dx < self.size[0] and 0 <= y + dy < self.size[1]]
        if moves:
            self.pos = list(random.choice(moves))
            return f"Moved to ({self.pos[0]},{self.pos[1]})"
        return "No moves available"

    def display(self):
        """Display grid with agent position."""
        for i in range(self.size[0]):
            row = ['V' if [i, j] == self.pos else self.grid[i][j] for j in range(self.size[1])]
            print(' '.join(row))
        print()

def main():
    agent = VacuumReflexAgent()
    print("Initial Grid:")
    agent.display()
    step = 0
    max_steps = 10000 # Increased to ensure full cleaning
    while not agent.is_room_clean() and step < max_steps:
        step += 1
        print(f"Step {step}: {agent.act()}")
        agent.display()

    if agent.is_room_clean():
        print(f"All clean! Total cleaned: {agent.cleaned} in {step} steps")
    else:
        print(f"Max steps ({max_steps}) reached. Room not fully cleaned.")

if __name__ == "__main__":
    main()
```
